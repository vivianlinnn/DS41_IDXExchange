```python
import pandas as pd

# Load
train = pd.read_csv("../processed_data/train_cleaned.csv")
test = pd.read_csv("../processed_data/test_cleaned.csv")

# ===== Add Features =====
for df in [train, test]:
    df["PPSF"] = df["ClosePrice"] / df["LivingArea"]
    df["Bed_Bath_Ratio"] = df["BedroomsTotal"] / df["BathroomsTotalInteger"]
    df["Property_Age"] = 2025 - df["YearBuilt"]

    df["CloseDate"] = pd.to_datetime(df["CloseDate"])
    df["Months_From_Dec_2025"] = (
        (2025 - df["CloseDate"].dt.year) * 12
        + (12 - df["CloseDate"].dt.month)
    )


train.to_csv("../processed_data/train_cleaned_fe.csv", index=False)
test.to_csv("../processed_data/test_cleaned_fe.csv", index=False)

print("Saved new feature-engineered files!")
```

## Additional Features
1. Sqft_Per_Bedroom


Why it matters:
Two homes can have identical total square footage but very different layouts:
	•	2000 sqft with 3 bedrooms → spacious rooms
	•	2000 sqft with 6 bedrooms → smaller, more crowded rooms





2. Lot_Utilization


Why it matters:
It helps differentiate property types and land value dynamics:
	•	High ratio → denser, urban-style housing
	•	Low ratio → larger lots, suburban or estate properties


```python

import numpy as np
train = pd.read_csv("../processed_data/train_cleaned_fe.csv")
test  = pd.read_csv("../processed_data/test_cleaned_fe.csv")

for df in [train, test]:
    
    # Avoid division by zero
    beds = df["BedroomsTotal"].replace(0, np.nan)
    lot  = df["LotSizeSquareFeet"].replace(0, np.nan)

   
    df["Sqft_Per_Bedroom"] = df["LivingArea"] / beds
    df["Lot_Utilization"] = df["LivingArea"] / lot


# Save back
train.to_csv("../processed_data/train_cleaned_fe.csv", index=False)
test.to_csv("../processed_data/test_cleaned_fe.csv", index=False)

print("Added Sqft_Per_Bedroom and Lot_Utilization ")
```
